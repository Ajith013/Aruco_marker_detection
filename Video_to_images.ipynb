```python
import cv2
```

```python
video_capture = cv2.VideoCapture('calibration_video/checkerboard.mp4')
return_val, image = video_capture.read()
count = 0
while return_val:
    if count%50==0:
        cv2.imwrite("calibration_images/frame%d.jpg" % count, image)
    return_val, image = video_capture.read()
    print('Read a new frame: ', return_val)
    count += 1
```
